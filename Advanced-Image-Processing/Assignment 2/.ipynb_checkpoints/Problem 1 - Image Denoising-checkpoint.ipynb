# Problem 1 - Image Denoising

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import matplotlib.gridspec as gridspec
from scipy.ndimage.filters import gaussian_filter
import scipy.stats as st
import helper as hp # contains helper functions
```

#### Loading Image

```python
img_org = cv.imread('./lighthouse2.bmp', cv.IMREAD_GRAYSCALE)
rows, cols = img_org.shape;
img_org = np.float32(img_org)

hp.show(img_org, 'Original Image')
```

#### Adding Gaussian noise

```python
noise = np.ones((rows,cols), np.float32)
noise_sigma = 10; var_z = noise_sigma**2;
noise = cv.randn(noise, 0, noise_sigma)
img_noise = img_org + noise;

hp.show(img_noise, 'Image with noise')
```

#### Finding Kernel with best MSE

```python
Ksizes = [3,7,11];
Ksigmas = [0.1, 1, 2, 4, 8];
best_size ,best_sigma, best_mse = None, None, 1e30;

print('ksize \t sigma \t    MSE')
for ksize in Ksizes:
    for ksigma in Ksigmas:
        ker = hp.gaussianKernel(ksize, ksigma)
        uy  = cv.filter2D(img_noise, cv.CV_32F, ker)
        mse = hp.getMSE(uy, img_org)

        if(mse < best_mse):
            best_size = ksize
            best_sigma = ksigma
            best_mse = mse
        
        plt.rcParams["figure.figsize"] = (5,8)
        title = 'size: {:d}, Sigma: {:.2f}'.format(ksize, ksigma)
        file_name = './q1_1/'+str(ksize) + '_' + str(ksigma) + '.png'
        plt.imshow(uy, cmap='gray'), plt.title(title), plt.xticks([]), plt.yticks([]);
        plt.savefig(file_name, dpi= 200, bbox_inches='tight', transparent=False)
        
        print('{:d} \t {:.2f} \t {:f}'.format(ksize, ksigma, mse))

print('\nKernel with best MSE => size: {:d}, sigma: {:.2f}'.format(best_size, best_sigma))
```

#### Comparing differnt Gaussian kernels

```python
im1 = cv.imread('./q1_1/3_0.1.png', cv.IMREAD_GRAYSCALE)
im2 = cv.imread('./q1_1/3_1.png', cv.IMREAD_GRAYSCALE)
im3 = cv.imread('./q1_1/7_4.png', cv.IMREAD_GRAYSCALE)
im4 = cv.imread('./q1_1/11_8.png', cv.IMREAD_GRAYSCALE)

plt.rcParams["figure.figsize"] = (10,7)
hp.subShow(2,2,1,im1)
hp.subShow(2,2,2,im2)
hp.subShow(2,2,3,im3)
hp.subShow(2,2,4,im4)
plt.subplots_adjust(wspace=0.00, hspace=0)
plt.savefig('q1_1/comp.jpg', dpi= 200, bbox_inches='tight', transparent=False)
plt.show()
```

## Part - 2: MMSE filter

```python
# Covolving with gaussian filter
ker = hp.gaussianKernel(best_size, best_sigma)
uy = cv.filter2D(img_noise, cv.CV_32F, ker)

# Computing Variance of the high pass image 
y1 = img_noise - uy
var_y1 = np.sum(y1*y1) / (rows*cols)

# Computing variance of noise in high pass image
w00 = ker[int(best_size/2), int(best_size/2)]
var_z1 = (1 - w00)**2 * var_z + (np.sum(ker*ker) - w00**2) * var_z

var_x1 = var_y1 - var_z1
x1 = (var_x1 / (var_x1 + var_z1))* y1;
x_hat_mmse = uy + x1;

plt.rcParams["figure.figsize"] = (10,10)
hp.subShow(1,2,1, uy, 'Low pass Image')
hp.subShow(1,2,2, x_hat_mmse, 'MMSE output')
plt.subplots_adjust(wspace=0.05, hspace=.01)
plt.savefig('mmse.jpg', dpi= 200, bbox_inches='tight', transparent=False)
plt.show()

print('Variance y1', var_y1)
print('Variance z1', var_z1)
print('MSE low pass', hp.getMSE(uy, img_org))
print('MSE MMSE', hp.getMSE(x_hat_mmse, img_org))
```

#### Displaying estimated high pass component 

```python
plt.rcParams["figure.figsize"] = (5,5)
x11 = x1 > 10
title = 'High pass component (MMSE)'
plt.imshow(x11, cmap='gray'), plt.title(title), plt.xticks([]), plt.yticks([]);
plt.savefig('mmse_hp.png', dpi= 200, bbox_inches='tight', transparent=False)
plt.show()
```

## Part - 3: Sure shrink

#### Defining helper functions

```python
def g(y1, t):
    M,N = y1.shape;
    res = np.zeros(y1.shape, np.float32)
    
    res[y1 < -t] = t
    res[y1 > t] = -t
    res[np.abs(y1) < t] = -y1[np.abs(y1) < t]
                
    return res

def getSURE(var, y1, t):
    M,N = y1.shape;
    g_y1 = g(y1, t)
    
    return M*N*var + np.sum(g_y1*g_y1) + 2 * var * (-1 * np.sum(np.abs(y1) < t)) 
```

#### Getting optimal threshold using SureShrink

```python
# Estimate best threshold 't'
best_t, best_sure = None, 1e30;
sures = []
t_list = np.linspace(0, 50, 200)

for t in t_list:
    sure = getSURE(var_z1, y1, t)
    
    if(best_sure > sure):
        best_sure = sure
        best_t = t
    sures.append(sure)


plt.rcParams["figure.figsize"] = (10,5)
plt.title('Estimating threshold using SureShrink')
plt.plot(t_list, sures)
plt.xlabel('t (threshold)')
plt.ylabel('SURE')
plt.savefig('./q1_1/3_sureComp.jpg', dpi= 200, bbox_inches='tight', transparent=False)
plt.show()

print('Best \'t\' value: {:f}, Best SURE: {:f}'.format(best_t, best_sure))
```

#### Verifying the estimated threshold from SureShrink

```python
vec_mse = []
around_best_t = np.linspace(best_t - 5, best_t +5, 20)
around_best_t[int(len(around_best_t)/2)] = best_t
print('t \t mse')
for t in around_best_t:
    zero_vec = np.zeros(y1.shape, np.float32)
    x1_sure = np.sign(y1) * np.maximum(zero_vec,(np.abs(y1) - t))
    x_hat_sure = uy + x1_sure;
    
    mse = hp.getMSE(x_hat_sure, img_org)
    vec_mse.append(mse)
    print('{:.2f} \t {:.4f}'.format(t, mse))

plt.plot(around_best_t, vec_mse)
plt.xlabel('t')
plt.ylabel('MSE')
```

#### Applying Shrinkage estimator

```python
zero_vec = np.zeros(y1.shape, np.float32)

x1_sure = np.sign(y1) * np.maximum(zero_vec,(np.abs(y1) - best_t))
x_hat_sure = uy + x1_sure;
x11_sure = x1_sure > 10
print('MSE: ', hp.getMSE(x_hat_sure, img_org))
plt.rcParams["figure.figsize"] = (10,7)
hp.subShow(2,2,1, uy, 'Low Pass Image');
hp.subShow(2,2,2, x_hat_sure, 'Shrinkage Estimator (t = 9.3)');
hp.subShow(2,2,3, x_hat_mmse, 'MMSE output')
hp.subShow(2,2,4, x11_sure, 'HP component (Shrinkage, t = 9.3)')
plt.subplots_adjust(wspace=0.05, hspace=.08)
plt.savefig('q1_1/3_shrink_comp.jpg', dpi= 200, bbox_inches='tight', transparent=False)
plt.show()
```

## Part - 4

#### Getting optimal Gaussian kernel for second scale

```python
Ksizes = [17,21,23,25,31];
Ksigmas = [4,8,12,16];
best_size2 ,best_sigma2, best_mse2 = None, None, 1e30;

print('ksize \t sigma \t    MSE')
for ksize in Ksizes:
    for ksigma in Ksigmas:
        
        ker = hp.gaussianKernel(ksize, ksigma)
        uy2 = cv.filter2D(uy, cv.CV_32F, ker)
        mse = hp.getMSE(uy2, img_org)

        if(mse < best_mse2):
            best_size2 = ksize
            best_sigma2 = ksigma
            best_mse2 = mse
            
        print('{:d} \t {:.2f} \t {:f}'.format(ksize, ksigma, mse))

print('\nKernel with best MSE => size: {:d}, sigma: {:.2f}'.format(best_size2, best_sigma2))
ker1 = hp.gaussianKernel(best_size, best_sigma)
ker2 = hp.gaussianKernel(best_size2, best_sigma2)
```

#### Comparing different Gaussian filters

```python
uy_new = cv.filter2D(uy, cv.CV_32F, ker1) 
uy_new2 = cv.filter2D(uy, cv.CV_32F, ker2)
ker3 = hp.gaussianKernel(31, 8);
uy_new3 = cv.filter2D(uy, cv.CV_32F, ker3)

print(hp.getMSE(uy_new, uy))
print(hp.getMSE(uy_new2, uy))
print(hp.getMSE(uy_new3, uy))

plt.rcParams["figure.figsize"] = (10,7)
hp.subShow(2,2,1, uy, 'Low pass first phase');
hp.subShow(2,2,2, uy_new, 'Applying kernel from first phase');
hp.subShow(2,2,3, uy_new2, 'Gaussian (size = 17, sig = 4)');
hp.subShow(2,2,4, uy_new3, 'Gaussian (size = 31, sig = 8)');
plt.subplots_adjust(wspace=0.05, hspace=.08)
plt.savefig('q1_1/same_filter.jpg', dpi= 200, bbox_inches='tight', transparent=False)
plt.show()
```

#### Estimating optimal threshold for second scale using SureShrink

```python
# best_size, best_sigma = 31, 8

uy2 = cv.filter2D(uy, cv.CV_32F, ker2)
y2 = uy - uy2

# Computing variance of noise in high pass of uy
var_uz = var_z * np.sum(ker1 * ker1)
w00 = ker2[int(best_size/2), int(best_size/2)]
var_z2 = (1 - w00)**2 * var_uz + (np.sum(ker2*ker2) - w00**2) * var_uz

# Estimate best threshold 't'
best_t, best_sure = None, 1e30;
sures = []


for t in np.linspace(0, 20, 100):
    sure = getSURE(var_z2, y2, t)

    if(best_sure > sure):
        best_sure = sure
        best_t = t
    sures.append(sure)

plt.rcParams["figure.figsize"] = (10,5)
plt.title('Estimating threshold using SureShrink')
plt.plot(np.linspace(0, 20, 100), sures)
plt.xlabel('t (threshold)')
plt.ylabel('SURE')
plt.savefig('./q1_1/4_sureComp.jpg', dpi= 200, bbox_inches='tight', transparent=False)
plt.show()

print('Best \'t\' value: {:f}, Best SURE: {:f}'.format(best_t, best_sure))
```

#### Applying Shrinkage estimator

```python
zero_vec = np.zeros(y2.shape, np.float32)

x2_sure = np.sign(y2) * np.maximum(zero_vec,(np.abs(y2) - best_t))
x22_sure = x2_sure > 10
x11_sure = x1_sure > 12
x_hat_multi = uy2 + x1_sure + x2_sure;

print('MSE: ', hp.getMSE(x_hat_multi, img_org))
plt.rcParams["figure.figsize"] = (10,7)
hp.subShow(2,2,1, x_hat_multi, 'Two scale denoising');
hp.subShow(2,2,2, x_hat_sure, 'One scale denoising');
hp.subShow(2,2,3, x11_sure, 'High pass first scale')
hp.subShow(2,2,4, x22_sure, 'High pass second scale')
plt.subplots_adjust(wspace=0.05, hspace=.08)
plt.savefig('q1_1/4_shrink_comp.jpg', dpi= 200, bbox_inches='tight', transparent=False)
plt.show()
```
