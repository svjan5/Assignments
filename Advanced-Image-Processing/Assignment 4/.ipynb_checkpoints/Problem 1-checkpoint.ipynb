```python
SI   = 5;                  # Color similarity
SX   = 6;                  # Spatial similarity
r    = 1.5;                # Spatial threshold (less than r pixels apart)
sNcut = 0.21;              # The smallest Ncut value (threshold) to keep partitioning
sArea = 40; 
```

```python
euc_fea = euclidean_distances(X_fea)
W_fea = np.exp(-euc_fea/(SI))

euc_pos = euclidean_distances(X_pos)
pos_mask = np.int8(euc_pos < r)
W_pos = pos_mask * np.exp(-euc_pos/(SX))

W = W_fea * W_pos
d = W.sum(axis=1)
D = np.diag(d)
```

```python
%%time
L = D - W
eigVal, eigVec = np.linalg.eig(L)
eigVal = eigVal.astype('float')
eigVec = eigVec.astype('float')
eigInd = np.argsort(eigVal)
```

```python
eig_vec2 = eigVec[eigInd[1]]
print(np.mean(eig_vec2), np.max(eig_vec2), np.min(eig_vec2))
```

```python
clust_ind = np.zeros((m*n,1), np.int8)
clust_ind = np.int8(eig_vec2 > 0.000001)

img_res = np.zeros((m,n,3), np.float32)
counter = 0;
for i in range(m):
    for j in range(n):
        img_res[i,j,clust_ind[counter]] = 1
        counter += 1

hp.cshow(img_res)
```
