```python
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from sklearn.metrics.pairwise import rbf_kernel
from sklearn.cluster import SpectralClustering
from sklearn.cluster import KMeans
```

#### Load Image

```python
img = cv.imread('Data.jpg', cv.IMREAD_GRAYSCALE)
rows, cols = img.shape
img = img < 200
```

#### Make Image sparse

```python
Mask = np.random.rand(rows, cols)
Mask = Mask > .9
img_sparse  = np.multiply(img, Mask)

plt.subplot(1,2,1), plt.imshow(img, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(img_sparse, cmap='gray'),  plt.xticks([]), plt.yticks([])
plt.show()
```

#### Generate Data points

```python
X = []
Y = []
for i in (range(rows)):
    for j in (range(cols)):
        if(img_sparse[i,j] == True):
                X.append(j)
                Y.append(i)
X.reverse()
# Y.reverse()  

plt.plot(X,Y, '.')
plt.show()
```

```python
d = {'X_coord': X, 'Y_coord': Y}
df = pd.DataFrame(data=d)
X = df.as_matrix()
m,n = X.shape;
```

#### Computing Similarity Matrix

```python
A = np.zeros((m, m), np.float32);
sigma = 5

A = rbf_kernel(X)
d = A.sum(axis = 1)
D = np.diag(d)
```

```python
X.shape
```

#### Computing Lagrangian Matrix

```python
L = D - A

eigVal, eigVec = np.linalg.eig(L)
eigVal = eigVal.astype('float')
eigVec = eigVec.astype('float')
eigInd = np.argsort(eigVal)
```

#### Applying K means

```python
k = 4
reqEigVec = eigVec[:,eigInd[range(k)]]
reqEigVal = eigVal[eigInd[range(k)]]
kmeans = KMeans(n_clusters=k, random_state=0).fit(reqEigVec)
clust_ind = kmeans.predict(reqEigVec)
df['Cluster'] = clust_ind
```

```python
reqEigVec.shape
```

```python
plt.scatter(reqEigVec[:,1], reqEigVec[:,2])
plt.show()
```

```python
plt.plot(range(m), eigVal)
plt.xlabel('k')
plt.ylabel('Eigen Value')
plt.show()
```
